```python
import pandas as pd
```

```python
# loading clean data for analysis
```

```python
df = pd.read_csv("../data_clean/Book1.csv")

print(df.to_string())
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# below graph shows the distribution of Age in the given data
```

```python
ax=sns.distplot(df['Age'], kde = False, color ='blue', bins = 30)
fig=ax.get_figure()
fig.savefig("../results/Age_Distibution.png")
```

```python
# below graph shows the distribution of Gripstrength in the given data
```

```python
ax=sns.distplot(df['Grip Strength'], kde = False, color ='green', bins = 30)
fig=ax.get_figure()
fig.savefig("../results/Grip_Strength_Distibution.png")
```

```python
#below graph shows how the gripstrength increases with age and decreases after certain age
```

```python
ax = df.plot.scatter(x='Age',y='Grip Strength')
fig=ax.get_figure()
fig.savefig("../results/Grip_Strength_Age_Relation.png")
```

```python

# below graph shows the relation between gripstrength and weight and how iut varies on weight.
```

```python
ax = df.plot.scatter(x='Weight',y='Grip Strength')
fig=ax.get_figure()
fig.savefig("../results/GripStrength_Weight_Relation.png")
```
